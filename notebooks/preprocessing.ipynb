```python
import pandas as pd
import csv
```

```python
# Encoding that was used for the dataset is ISO-8859-15

o3_2023 = pd.read_csv("../data/raw/O3/2023_O3.csv", encoding = 'ISO-8859-15', sep =';', skiprows = 9)
o3_2024_01 = pd.read_csv("../data/raw/O3/2024_01_O3.csv", encoding = 'ISO-8859-15', sep =';', skiprows = 9)
o3_2024_02 = pd.read_csv("../data/raw/O3/2024_02_O3.csv", encoding = 'ISO-8859-15', sep =';', skiprows = 9)
o3_2024_03 = pd.read_csv("../data/raw/O3/2024_03_O3.csv", encoding = 'ISO-8859-15', sep =';', skiprows = 9)
o3_2024_04 = pd.read_csv("../data/raw/O3/2024_04_O3.csv", encoding = 'ISO-8859-15', sep =';', skiprows = 9)
o3_2024_05 = pd.read_csv("../data/raw/O3/2024_05_O3.csv", encoding = 'ISO-8859-15', sep =';', skiprows = 9)
o3_2024_06 = pd.read_csv("../data/raw/O3/2024_06_O3.csv", encoding = 'ISO-8859-15', sep =';', skiprows = 9)
o3_2024_07 = pd.read_csv("../data/raw/O3/2024_07_O3.csv", encoding = 'ISO-8859-15', sep =';', skiprows = 9)
```

```python
o3_2023.head()
```

```python
# Utrecht-Kardinaal de Jongweg id : NL10636
# Utrecht-Griftpark id: NL10643
```

```python
def filter_data(data):
    return data[['Component', 'Einddatumtijd', 'NL10643']]
                

filtered_o3_2023 = filter_data(o3_2023)
filtered_o3_2024_01 = filter_data(o3_2024_01)
filtered_o3_2024_02 = filter_data(o3_2024_02)
filtered_o3_2024_03 = filter_data(o3_2024_03)
filtered_o3_2024_04 = filter_data(o3_2024_04)
filtered_o3_2024_05 = filter_data(o3_2024_05)
filtered_o3_2024_06 = filter_data(o3_2024_06)
filtered_o3_2024_07 = filter_data(o3_2024_07)
```

```python
filtered_o3 = pd.concat([filtered_o3_2023, filtered_o3_2024_01, filtered_o3_2024_02,
                         filtered_o3_2024_03, filtered_o3_2024_04, filtered_o3_2024_05,
                         filtered_o3_2024_06, filtered_o3_2024_07], axis = 0)
```

```python
filtered_o3 = filtered_o3.reset_index(drop = True)
filtered_o3
```

```python
filtered_o3.to_csv('../data/processed/filtered_03.csv', index = False)
```
